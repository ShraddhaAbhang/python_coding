```python
# Create a pivot table with 'car_name' as index and 'price($)' as values, aggregating by mean
print(pd.pivot_table(df, values='price($)', index='car_name', aggfunc=np.mean))
```

```python
# Melt the DataFrame to transform it from wide format to long format
print(pd.melt(df, id_vars=['car_name'], value_vars=['price($)', 'meter_reading']))
```

```python
# Merge two DataFrames on a common column
# Inner join: only include rows with keys present in both DataFrames
df1 = pd.DataFrame({'car_name': ['Lamborghini', 'Porsch', 'Honda City'], 'horsepower': [100, 150, 200]})
df_merged = pd.merge(df, df1, on='car_name', how='inner')
df_merged
```

```python
# Left join: include all rows from the left DataFrame, with matching rows from the right DataFrame
df_left = pd.merge(df, df1, on='car_name', how='left')
print("\nLeft join (all rows from left DataFrame):")
df_left
```

```python
# Right join: include all rows from the right DataFrame, with matching rows from the left DataFrame
df_right = pd.merge(df, df1, on='car_name', how='right')
print("\nRight join (all rows from right DataFrame):")
df_right
```

```python
# Outer join: include all rows from both DataFrames, with matching rows where available
df_outer = pd.merge(df, df1, on='car_name', how='outer')
print("\nOuter join (all rows from both DataFrames):")
df_outer
```

```python
# Concatenate two DataFrames along rows
df3 = pd.DataFrame({'car_name': ['Car D', 'Car E'], 'price($)': [300000, 400000]})
df_concat = pd.concat([df, df3], ignore_index=True)
df_concat
```

```python
# Rename columns in the DataFrame
df_renamed = df.rename(columns={'price($)': 'price_in_dollars', 'meter_reading': 'mileage'})
df_renamed
```
