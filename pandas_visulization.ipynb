```python
bank_data_df = pd.read_csv("bank.csv")
bank_data_df.head()
```

```python
# Line graph
plt.figure(figsize=(10,6))
plt.plot(bank_data_df["duration"], bank_data_df["balance"])
plt.title("Printing Line graph of duration vs balance")
plt.xlabel("duration")
plt.ylabel("balance")
plt.grid(True)
plt.show()
```

```python
# Histogram
plt.figure(figsize=(10,6))
plt.hist(bank_data_df["age"], bins=100)
        #  , bank_data_df["balance"])
plt.title("Printing Line graph of duration vs balance")

# plt.xlabel("duration")
# plt.ylabel("balance")
plt.grid(True)
plt.show()
```

```python
# Scatter 
plt.figure(figsize=(10,6))
plt.scatter(bank_data_df["age"], bank_data_df["balance"], c="pink")
plt.title("Printing Line graph of duration vs balance")
plt.xlabel("age")
plt.ylabel("balance")
plt.grid(True)
plt.show()
```
